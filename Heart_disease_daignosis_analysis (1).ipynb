```python
#None of the missing values found above
#Using function below to convert numerical into categorical value

def heart_disease(value):
    if value == 0:
        return 'absence'
    else:
        return 'presence'

#adding new column to dataset of converted data
df['heart_diseases'] = df['target'].apply(heart_disease)
df.head()



# Define the new order of columns
new_column_order = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target', 'heart_diseases']

# Reorder the DataFrame
df = df[new_column_order]

# Display the DataFrame
df.head()
```

```python
# To find the minimum,maximum and average of the population age using statistical analysis
import numpy as np
min_age = df['age'].min()
max_age = df['age'].max()
mean_age = df['age'].mean()
print(f"The minimum age is {min_age}")
print(f"The maximum age is {max_age}")
print(f"The average age is {np.round(mean_age, 1)}")
```

```python
# To divide the population age in different categories

df = pd.read_csv(r'C:\Users\Mansi\Downloads\Heart Disease data.csv', encoding= 'unicode_escape')
young_age = df[(df['age']>=29) & (df['age']<40)]
middle_age = df[(df['age']>=40) & (df['age']<55)]
old_age = df[(df['age']>55)]
print(f"Number of young age people = {len(young_age)}")
print(f"Number of middle age people = {len(middle_age)}")
print(f"Number of old age people = {len(old_age)}")
```

```python
# Bar plot using matplotlib and seaborn for different categories of population age
import matplotlib.pyplot as plt
category = ['young_age','middle_age', 'old_age']
count_age_category = [len(young_age), len(middle_age), len(old_age)]
plt.bar(category, count_age_category, color=['yellow', 'orange', 'red'])
plt.title('Age category')
plt.xlabel('Age range')
plt.ylabel('count')
plt.show()
```

```python
#Based on above observation it can be concluded that old age population is more than middle age range and young age population is the least
```

```python
#converting numerical data into categorical data

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 

def age_range(row):
    '''converting population age into range of age'''
    if row >= 29 and row < 40: 
        return 'youngage'
    elif row >= 40 and row < 55:
        return 'middleage'
    else:
        return 'oldage'

#applying converted data into our dataset
df['ageRange'] = df['age'].apply(age_range)
df.head()

# Define the new order of columns
new_column_order = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target', 'heart_diseases', 'ageRange']

# Reorder the DataFrame
df = df[new_column_order]

# Display the DataFrame
df.head()
```

```python
#converting numerical data into categorical data
def sex(row):
    if row == 1:
        return 'male'
    else:
        return 'female' 

#applying converted data into our dataset
df['gender'] = df['sex'].apply(sex)
df.head() 

# Define the new order of columns
new_column_order = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target', 'heart_diseases', 'ageRange', 'gender']

# Reorder the DataFrame
df = df[new_column_order]

# Display the DataFrame
df.head()

```
